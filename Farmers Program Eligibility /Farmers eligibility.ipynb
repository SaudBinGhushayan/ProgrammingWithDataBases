## Checking Eligibility users for Farmers Program
__Rules To Be Eligible To The Program__ :
1. Salary must be below 10,000
2. Age must be between 20 and 70 years old
3. Soical Status must be married 
4. He must have more than 1 child
5. He must Have A Citizan Account
6. He shouldn't have any extra income 
7. His account balance must be below 100,000
8. He must have below 50,000 sheeps

```python
#Importing Libraries
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sb
```

```python
#Reading the csv File
df = pd.read_csv('Eligibility.csv')
```

```python
#Viewing the dataset
df
```

```python
#General information of the dataset
df.info()
```

```python
#Getting statistical info
df.loc[:,'salary':].describe()
```

```python
#Checking for duplicates
df.duplicated().any().sum()
```

```python
#Dropping duplicates if any found
df.drop_duplicates(inplace = True)
```

```python
#Verifying that duplicates has been dropped if fouund
df.duplicated().any().sum()
```

```python
#getting the accepted applicants
Accepted = ((df['salary'] < 10000) & df['age'].between(20,70) & (df['status'] =='M') &
           (df['child']>=2) & (df['citAcc']==True) & (df['extraIncome']==False) & (df['balance']<=100000)&
           (df['sheep']<=50000))
df[Accepted]
```

```python
#getting rejected applicants
Not_Accepted = ((df['salary'] > 10000) | ((df['age']<20) | (df['age']>70)) | (df['status'] =='S') |
                (df['child']<=2) | (df['citAcc']==False) | (df['extraIncome']==True) | (df['balance']>100000)|
                (df['sheep']>50000))
df[Not_Accepted]
```

# Visualization

## Comparing Accepted Applicants to Rejected

```python
#A pie chart showing a percentage of accepted and Rejected people for the program
nb_Accepted , nb_not_Accepted = df[Accepted]['id'].count() ,  df[Not_Accepted]['id'].count()
percent_chart = pd.DataFrame([nb_Accepted, nb_not_Accepted],
                        index=['Acccepted','Rejected'] ,
                        columns = ['Percentage of Accepted and Rejected people\n pie chart'])
#now plot the pie chart
percent_chart.plot(kind='pie', subplots=True, figsize=(6,6),autopct='%1.1f%%');
```

## How many Applicants Rejected for the children condition?

```python
#Gettign number of Rejected Applicants
df[Not_Accepted]['id'].count()
```

```python
#Plotting a bar chart that take the number of children for rejected applicants
base_color = sb.color_palette()[0]  
graph = sb.catplot(data=df[Not_Accepted], x='child',                 
                   kind='count',color = base_color)
graph.set_axis_labels('Number of Children', 'Rejected Applicants')
graph.fig.suptitle('Comparing Applicants Children',
                   y=1.07, fontsize=20, fontweight='bold');
```
